```python
# we have a multicollinearity problem as predicted (lots of size/weight and engine performance variables are heavily correlated)
# meaning that coefficients are biased and we cannot trust results above
```

```python
# eliminate the problem by dropping the variables that cause the problem 
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# compute the vif for all given features
def compute_vif(considered_features):
    
    X = data_t[considered_features]
    # the calculation of variance inflation requires a constant
    X['intercept'] = 1
    
    # create dataframe to store vif values
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif = vif[vif['Variable']!='intercept']
    return vif
```

```python
# features to consider removing
considered_features = ['fuel-type_diesel',  
       'aspiration_turbo',  'num-of-doors_two',
       'body-style_convertible', 'body-style_hatchback', 
       'body-style_wagon', 'drive-wheels_4wd', 'drive-wheels_fwd',
       'drive-wheels_rwd', 'engine-location_rear',
       'engine-type_dohc', 'engine-type_dohcv', 'engine-type_l',
        'engine-type_ohcf', 'engine-type_ohcv',
       'engine-type_rotor', 
        'fuel-system_1bbl', 'fuel-system_2bbl',
       'fuel-system_4bbl', 'fuel-system_idi', 'fuel-system_mfi',
        'fuel-system_spdi', 'fuel-system_spfi',
       'wheel-base', 'length', 'width', 'height',  'engine-size',
       'compression-ratio', 'city-mpg', 'highway-mpg', 'num-of-cylinders',
       'bore', 'stroke', 'horsepower', 'peak-rpm', 
       'normalized-losses']

# compute vif 
compute_vif(considered_features).sort_values('VIF', ascending=False)
```

```python
def myplot(score,coeff,labels=X.columns):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley, c=data.groupby(['make']).ngroup())
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))
    plt.grid()

#Call the function. Use only the 2 PCs.
x_new = data_pca 
plt.figure(figsize=(20,15 ))
myplot(x_new[:,0:2],np.transpose(pca.components_[0:2, :]))
plt.show()
```

```python
# create train-test split
from sklearn.model_selection import train_test_split
X = data_t[['fuel-type_diesel', 'fuel-type_gas', 'aspiration_std',
       'aspiration_turbo', 'num-of-doors_four', 'num-of-doors_two',
       'body-style_convertible', 'body-style_hatchback', 'body-style_sedan',
       'body-style_wagon', 'drive-wheels_4wd', 'drive-wheels_fwd',
       'drive-wheels_rwd', 'engine-location_front', 'engine-location_rear',
       'engine-type_dohc', 'engine-type_dohcv', 'engine-type_l',
       'engine-type_ohc', 'engine-type_ohcf', 'engine-type_ohcv',
       'engine-type_rotor', 'fuel-system_1bbl', 'fuel-system_2bbl',
       'fuel-system_4bbl', 'fuel-system_idi', 'fuel-system_mfi',
       'fuel-system_mpfi', 'fuel-system_spdi', 'fuel-system_spfi',
       'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size',
       'compression-ratio', 'city-mpg', 'highway-mpg', 'num-of-cylinders',
       'bore', 'stroke', 'horsepower', 'peak-rpm', 
       'normalized-losses']]
y = data_t['price']
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1234, test_size = 0.33)
```

```python
# import the rfr, RMSE
from sklearn.ensemble import RandomForestRegressor 
from sklearn.metrics import mean_squared_error
import math

# we have lots of features, let's be generous with max_depth to begin with
rfr = RandomForestRegressor(max_depth=3, random_state=0)

model_rfr = rfr.fit(X_train, y_train)
predictions_rfr = model_rfr.predict(X_test)

print(mean_squared_error(y_test, predictions_rfr))
print(math.sqrt(mean_squared_error(y_test, predictions_rfr)))

fig, ax = plt.subplots()
ax.scatter(predictions_rfr, y_test)
ax.axline([0, 0], [1, 1])
ax.set_xlim(8.5,11)
ax.set_ylim(8.5,11)
```

```python
# model looks pretty accurate save for outliers at the high price point and the low price point
# finetuning the hyperparameters. 

# as I don't know how gridsearch works regressors we'll keep it simple.
# we'll keep it simple and limit ourselves to max_depth for now. 
# we'll leave maximum number of features to be included at each node split alone as we already have a lot correlation in the dataset, we might even decrease it downwards.
# we can also adjust  the number of estimators - more trees can increase accurarcy but take more compute (takes longer)
'''
max_depth 1 20
n_estimators int, default=100
max_features{â€œsqrtâ€, â€œlog2â€, None}, int or float, default=â€sqrtâ€'''

md = [i for i in range(2,21,2)]
e = [i for i in range(20,201,20)]
mf = [i for i in ['sqrt','log2','auto']]

# manually made grid (ugly nested loops)
grid = []
for i in range(len(md)):
    for j in range(len(e)):
        for k in range(len(mf)):
            grid.append((md[i],e[j],mf[k]))

results = []
for i in grid:
    rfr = RandomForestRegressor(max_depth=i[0], 
                                random_state=0,
                                n_estimators=i[1],
                                max_features=i[2])

    model_rfr = rfr.fit(X_train, y_train)
    predictions_rfr = model_rfr.predict(X_test)

    rmse = math.sqrt(mean_squared_error(y_test, predictions_rfr))
    
    results.append([i,rmse])
```

```python
pd.DataFrame(np.array(results)).sort_values(1, ascending=True)
```

```python
# best parameters (12, 60, log2)
rfr = RandomForestRegressor(max_depth=12, 
                                random_state=0,
                                n_estimators=60,
                                max_features='log2')

model_rfr = rfr.fit(X_train, y_train)
predictions_rfr = model_rfr.predict(X_test)

rmse = math.sqrt(mean_squared_error(y_test, predictions_rfr))
    
print(rmse)

fig, ax = plt.subplots()
ax.scatter(predictions_rfr, y_test)
ax.axline([0, 0], [1, 1])
ax.set_xlim(8.5,11)
ax.set_ylim(8.5,11)
plt.show()
```
